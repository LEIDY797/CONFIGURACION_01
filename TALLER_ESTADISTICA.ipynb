```python
# Librer√≠as necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, chi2_contingency, f_oneway
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Cargar archivo y limpiar columnas
df = pd.read_csv('/content/dynamic_pricing.csv')
df.columns = df.columns.str.strip()

# PUNTO 5: Correlaciones dos a dos con Costo Hist√≥rico

# Variables involucradas
variables = {
    'Number_of_Riders': 'N√∫mero de Pasajeros',
    'Average_Ratings': 'Calificaci√≥n Promedio',
    'Expected_Ride_Duration': 'Duraci√≥n Esperada del Viaje'
}
target = 'Historical_Cost_of_Ride'

# Scatterplots
for var, label in variables.items():
    sns.scatterplot(x=df[var], y=df[target])
    plt.title(f'{label} vs Costo Hist√≥rico del Viaje')
    plt.xlabel(label)
    plt.ylabel('Costo Hist√≥rico del Viaje')
    plt.show()

# Correlaci√≥n de Pearson
print(" Correlaciones de Pearson con Costo Hist√≥rico del Viaje:\n")
for var, label in variables.items():
    r, p = pearsonr(df[var], df[target])
    fuerza = "fuerte" if abs(r) > 0.7 else "media" if abs(r) > 0.4 else "d√©bil"
    print(f"- {label} ‚Üî Costo Hist√≥rico: r = {r:.3f}, p-valor = {p:.4f} ‚Üí Correlaci√≥n {fuerza}")

# PUNTO 6: Tablas de contingencia y prueba Chi-Cuadrado

print("\n An√°lisis de asociaci√≥n entre variables categ√≥ricas (Chi-Cuadrado):\n")

cat_pairs = [
    ('Customer_Loyalty_Status', 'Time_of_Booking'),
    ('Customer_Loyalty_Status', 'Vehicle_Type'),
    ('Time_of_Booking', 'Vehicle_Type')
]

for var1, var2 in cat_pairs:
    tabla = pd.crosstab(df[var1], df[var2])
    print(f"\n Tabla de contingencia: {var1} vs {var2}\n")
    print(tabla)

    chi2, p, dof, expected = chi2_contingency(tabla)
    conclusion = "hay asociaci√≥n significativa" if p < 0.05 else "no hay asociaci√≥n significativa"
    print(f"Resultado prueba Chi-cuadrado: œá¬≤ = {chi2:.2f}, p-valor = {p:.4f} ‚Üí {conclusion}")

#  PUNTO 7: Boxplots y ANOVA seg√∫n Fidelidad del Cliente

grupo = 'Customer_Loyalty_Status'
variables_anova = {
    'Number_of_Riders': 'N√∫mero de Pasajeros',
    'Average_Ratings': 'Calificaci√≥n Promedio',
    'Expected_Ride_Duration': 'Duraci√≥n Esperada del Viaje'
}

for var, label in variables_anova.items():
    sns.boxplot(x=df[grupo], y=df[var], palette="Set2")
    plt.title(f'{label} seg√∫n Fidelidad del Cliente')
    plt.xlabel('Fidelidad del Cliente')
    plt.ylabel(label)
    plt.show()

    # ANOVA
    grupos = [grupo[var].values for name, grupo in df.groupby(grupo)]
    f_stat, p_val = f_oneway(*grupos)
    decision = "diferencias significativas" if p_val < 0.05 else "no hay diferencias significativas"
    print(f"ANOVA para {label} seg√∫n Fidelidad del Cliente: F = {f_stat:.2f}, p-valor = {p_val:.4f} ‚Üí {decision}")

    # Si hay diferencias significativas, hacer prueba Post Hoc
    if p_val < 0.05:
        posthoc = pairwise_tukeyhsd(endog=df[var], groups=df[grupo], alpha=0.05)
        print("\n Resultado prueba Post Hoc (Tukey HSD):")
        print(posthoc.summary())
```

```python
# PUNTO REGRESI√ìN LOG√çSTICA

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import numpy as np

# 1. Crear variable dummy: 0 = regular (< 4), 1 = excelente (>= 4)
df['Servicio_Excelente'] = (df['Average_Ratings'] >= 4).astype(int)

# 2. Codificar variables categ√≥ricas con get_dummies
df_encoded = pd.get_dummies(df[['Location_Category', 'Customer_Loyalty_Status', 'Time_of_Booking', 'Vehicle_Type']], drop_first=True)

# 3. Unir con la variable objetivo
X = df_encoded
y = df['Servicio_Excelente']

# 4. Dividir datos (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Ajustar modelo de regresi√≥n log√≠stica
modelo_log = LogisticRegression(max_iter=1000)
modelo_log.fit(X_train, y_train)

# 6. Obtener probabilidades y predicciones
y_pred = modelo_log.predict(X_test)
y_prob = modelo_log.predict_proba(X_test)

# 7. Interpretaci√≥n de los coeficientes (Odds Ratios)
odds_ratios = pd.DataFrame({
    'Variable': X.columns,
    'Odds Ratio': np.exp(modelo_log.coef_[0]),
    'Coeficiente': modelo_log.coef_[0]
})

print("\nüìä Odds Ratios e interpretaci√≥n:")
print(odds_ratios.sort_values(by='Odds Ratio', ascending=False))

# 8. Validaci√≥n del modelo
print("\nüìà Validaci√≥n del modelo:")
print(classification_report(y_test, y_pred, target_names=["Regular", "Excelente"]))
print("Accuracy del modelo:", accuracy_score(y_test, y_pred))

# 9. Matriz de confusi√≥n
matriz_conf = confusion_matrix(y_test, y_pred)
print("\nMatriz de Confusi√≥n:")
print(matriz_conf)

sns.heatmap(matriz_conf, annot=True, fmt="d", cmap="Blues", xticklabels=["Regular", "Excelente"], yticklabels=["Regular", "Excelente"])
plt.title("Matriz de Confusi√≥n")
plt.xlabel("Predicci√≥n")
plt.ylabel("Real")
plt.show()
```
